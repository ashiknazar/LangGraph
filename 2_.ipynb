### The 4 Things You Must Understand

___

#### 1. State (Everything revolves around this)

- State is just a shared dictionary that flows through your system.

```python
state = {
    "question": "What is GDP?",
    "answer": "",
    "valid": False
}
```

- Every step reads and updates this.

____

#### 2. Node (A single step)

A node is just a function.

```python
def generate_answer(state):
    state["answer"] = "some output"
    return state
```
- That’s it. No magic.

___

#### 3.Edge (What runs next)

Edges define flow:
- Go from A → B
- Or decide dynamically

___
#### 4.Graph (The full system)

You connect nodes into a graph.

___
___

## The Important Part — Control Flow

#### Linear Flow (basic)

```bash
A → B → C → END
```

___

#### Conditional Flow (decision)

```bash
A → B → (if good → END)
         (if bad → retry)
```

___

#### Loop (core idea)

```bash
Generate → Check → Generate → Check → END
```
- This is how you build self-correcting AI
